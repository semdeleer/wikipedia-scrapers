```python
from requests import Session

root_url = "https://country-leaders.onrender.com/"
status_url = "https://country-leaders.onrender.com/status/"
countries_url = "https://country-leaders.onrender.com/countries/"
cookie_url = "https://country-leaders.onrender.com/cookie/"
leaders_url = "https://country-leaders.onrender.com/leaders/"

def session_request(url: str, session: Session):
    response = session.get(url)
    response_json = response.json()
    status_response = session.get(status_url)

    if status_response.status_code == 200:
        return response_json
    else:
        print(status_response.status_code)
        return None

def session_loop_request(cookie: str, url: str):
    with Session() as session:
        session_request(cookie, session)  # Ignoring the return value for the cookie request
        country_list = session_request(url, session)
        
        if country_list:
            return country_list
    
def session_loop_request_param(cookie: str, url: str, country_code):
    leadersl={}
    for i in country_code:
        leaders_country_url = f"{url}?country={i}"

        with Session() as session:
            session_request(cookie, session)
            leadersl[i] = session_request(leaders_country_url, session)

            
    return leadersl


session_loop_request_param(cookie_url, leaders_url, session_loop_request(cookie_url, countries_url))
```

```python
def get_leaders():
    return session_loop_request_param(cookie_url, leaders_url, session_loop_request(cookie_url, countries_url))

get_leaders()
```

```python
def wiki_leader():
    leaderInfo = get_leaders()
    wikiLeaders = []
    for key, value in leaderInfo.items():
        for i in value:
            wikiLeaders.append(i["wikipedia_url"])

    return wikiLeaders
                
wiki_leader()
```

```python
def first_paraf():
    paragrapf = []
    ids = 1
    for wiki in wiki_leader():
        url = wiki
        r = requests.get(url)
        soup = BeautifulSoup(r.content, "html")
        for id, elem in enumerate(soup.find_all("p")):
                if id == ids and elem.text == '\n':
                    ids += 1
                if id == ids and elem.text != '\n':
                    paragrapf.append(elem.text)
                    break
    return paragrapf

first_paraf()
```

```python
def paraf_leader():
    dict_lead = get_leaders()
    dict_leader = {}
    for key, values in dict_lead.items():
        dict_leader = [{'first_name': person['first_name'], 'last_name': person['last_name']} for people in dict_lead.values() for person in people]
    list_paraf = first_paraf()
    lcount = 0
    for i in dict_leader:
        i["paraf"] = list_paraf[lcount]
        lcount += 1
    return dict_leader

paraf_leader()
```

```python
import json
def write_json():
    dictl = paraf_leader()
    brackt = re.compile(r'\[.*?\]')
    for i in dictl:
        i["paraf"] = re.sub(brackt, '', i["paraf"])

    
    with open("leadesparaf.json", "w", encoding='utf-8') as outfile:
        json.dump(dictl, outfile, indent=4, ensure_ascii=False)
    
    

write_json()
```

```python
def write_csv():
    #You need to add encoding utf8 after it because the json file is written with this param
    with open('leadesparaf.json', 'r', encoding='utf-8') as json_file:
        jsondata = json.load(json_file)
    
    #the isinstance check if a object is a specefic class
    if isinstance(jsondata, list) and jsondata:
        keys = jsondata[0].keys()  # Assuming all dictionaries in the list have the same keys

        with open('leadesparafcsv.csv', 'w', newline='', encoding='utf-8') as outfile:
            csv_writer = csv.writer(outfile)
            #the keys from the json file become the headers
            csv_writer.writerow(keys)

            for item in jsondata:
                #The values become the rows of the csv file
                csv_writer.writerow(item.values())

write_csv()
    
```
